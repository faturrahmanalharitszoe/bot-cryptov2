## Langkah 6: Download Model yang Sudah Dilatih
Setelah training selesai 100%, file model ber-ekstensi `.pt` akan tersimpan di folder `models/saved/`. Jalankan script ini untuk mengunduhnya ke komputer Anda.

```python
from google.colab import files
import os
import glob

saved_models = glob.glob('models/saved/*.pt')
if saved_models:
    # Ambil model terbaru
    latest_model = max(saved_models, key=os.path.getctime)
    print(f"Mengunduh: {latest_model}")
    files.download(latest_model)
else:
    print("Belum ada model yang tersimpan. Pastikan training selesai tanpa error.")
```
